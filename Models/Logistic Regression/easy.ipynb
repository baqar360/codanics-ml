```python
# Step 1: Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Step 2: Create simple dataset
data = {
    'Hours': [1, 2, 3, 4, 5, 6],
    'Marks': [20, 40, 50, 65, 80, 100]
}
df = pd.DataFrame(data)

# Step 3: Split into X and y
X = df[['Hours']]  # Input (must be 2D)
y = df['Marks']    # Output (1D)

# Step 4: Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 5: Train the model
model = LinearRegression()
model.fit(X_train, y_train)

# Step 6: Get slope and intercept
print("Slope (m):", model.coef_[0])
print("Intercept (b):", model.intercept_)

# Step 7: Predict
y_pred = model.predict(X_test)
print("Predictions:", y_pred)

# Step 8: Evaluate
print("MSE:", mean_squared_error(y_test, y_pred))
print("MAE:", mean_absolute_error(y_test, y_pred))

# Step 9: Plot
plt.scatter(X, y, color='blue', label='Actual')
plt.plot(X, model.predict(X), color='red', label='Predicted Line')
plt.xlabel('Hours Studied')
plt.ylabel('Marks Scored')
plt.title('Linear Regression Line')
plt.legend()
plt.grid(True)
plt.show()
```
